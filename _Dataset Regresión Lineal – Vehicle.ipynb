```python
# Importar bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Datos (Price, Kms_driven)
data1 = {
    'Price' : [50000,    95000,    250000,    150000,    40000,    50000,    60000,    43000,    65000,    75000,    95000,    70000,    100000,    60000,    175000,    1250000,    850000,    1190000,    600000,    390000,    450000,    850000,    550000,    250000,    1650000,    1195000,    240000],
    'Kms_driven' : [100000,    100000,    35000,    120000,    40000,    120000,    40000,    100000,    70000,    57000,    50000,    80000,    80000,    70000,    29173,    78000,    119000,    175900,    100000,    33000,    141000,    25000,    25000,    46000,    25000,    5000,    120000]
}
```

```python
# Convertir los datos en un DataFrame
df = pd.DataFrame(data1)

# Variables independientes (Kms_driven) y dependiente (Price)
X = df[['Kms_driven']]
y = df['Price']

# Crear un modelo de regresion lineal
modelo = LinearRegression()

# Ajustar el modelo de datos
modelo.fit(X, y)

# Coeficientes de la regresion
beta_0 = modelo.intercept_
beta_1 = modelo.coef_ 

# Mostrar los coeficientes
print(f"Intercepto (β0): {beta_0}")
print(f"Coeficiente (Kms_driven, β1): {beta_1}")

# Visualizar la regresión 
plt.scatter(df['Kms_driven'], y, color= "blue", label= "Datos Reales")
plt.plot(df['Kms_driven'], modelo.predict(X), color='red', label='Línea de Regresión')
plt.title('Regresión Lineal: Price vs Kms_driven')
plt.xlabel('Kms_driven')
plt.ylabel('Price')
plt.legend()
plt.show()
```
